```python
Solution

### Q1. What are the three measures of central tendency?

The three measures of central tendency are:
1. **Mean**
2. **Median**
3. **Mode**

### Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?

- **Mean**: The mean, often referred to as the average, is the sum of all values in a dataset divided by the number of values. It provides a measure of the central point but can be influenced by outliers.
  
- **Median**: The median is the middle value when the data is ordered from smallest to largest. If there is an even number of observations, the median is the average of the two middle values. The median is less affected by outliers and skewed data.

- **Mode**: The mode is the value that appears most frequently in a dataset. A dataset can have one mode, more than one mode, or no mode at all if all values are unique. The mode is useful for categorical data to identify the most common category.

These measures are used to provide different perspectives on the central tendency of a dataset, helping to understand its distribution and central point.

### Q3. Measure the three measures of central tendency for the given height data:
\[178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5\]

Let's calculate these:

- **Mean**:
\[ \text{Mean} = \frac{\sum \text{height}}{\text{number of heights}} = \frac{178 + 177 + 176 + 177 + 178.2 + 178 + 175 + 179 + 180 + 175 + 178.9 + 176.2 + 177 + 172.5 + 178 + 176.5}{16} \]

- **Median**: Arrange the data in ascending order and find the middle value(s).

- **Mode**: Identify the most frequently occurring value(s).

Let's do the calculations.

### Q4. Find the standard deviation for the given data:
\[178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5\]

The standard deviation measures the amount of variation or dispersion of a set of values. The formula for standard deviation (\( \sigma \)) is:
\[ \sigma = \sqrt{\frac{1}{N}\sum_{i=1}^{N}(x_i - \mu)^2} \]
where \( \mu \) is the mean of the data and \( x_i \) are the individual data points.

Let's calculate this.

### Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.

- **Range**: The range is the difference between the highest and lowest values in a dataset. It gives a quick sense of the spread but can be affected by outliers.

- **Variance**: Variance measures the average squared deviation of each number from the mean. It gives an idea of how data points spread out from the mean but in squared units.

- **Standard Deviation**: Standard deviation is the square root of the variance. It provides a measure of spread in the same units as the data, making it easier to interpret.

Example:
Consider the dataset \[4, 8, 6, 5, 3, 7\].
- **Range**: \(8 - 3 = 5\)
- **Variance**: First, find the mean \(\mu = 5.5\). Then calculate the average of the squared differences from the mean.
- **Standard Deviation**: The square root of the variance.

### Q6. What is a Venn diagram?

A Venn diagram is a diagram that shows all possible logical relations between a finite collection of different sets. Each set is represented by a circle, and the overlapping areas represent the intersections of the sets.

### Q7. For the two given sets \(A = \{2,3,4,5,6,7\}\) and \(B = \{0,2,6,8,10\}\), find:
(i) \(A \cap B\) (Intersection)
(ii) \(A \cup B\) (Union)

- (i) \(A \cap B\) is the set of elements that are in both \(A\) and \(B\).
- (ii) \(A \cup B\) is the set of elements that are in either \(A\) or \(B\) or both.

### Q8. What do you understand about skewness in data?

Skewness refers to the asymmetry in the distribution of data. If the data is not symmetrically distributed, it is said to be skewed. Skewness can be:

- **Positive skew (right-skewed)**: The right tail is longer; the mean is greater than the median.
- **Negative skew (left-skewed)**: The left tail is longer; the mean is less than the median.

### Q9. If a data is right skewed then what will be the position of median with respect to mean?

If a data set is right skewed (positively skewed), the median will be less than the mean.

### Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?

- **Covariance**: Covariance measures the directional relationship between two variables. A positive covariance indicates that the variables tend to move in the same direction, while a negative covariance indicates they move in opposite directions.

- **Correlation**: Correlation quantifies the strength and direction of the relationship between two variables. It is a standardized form of covariance that is dimensionless and ranges from -1 to 1. 

Covariance and correlation are used to understand and quantify the degree to which two variables are related. Correlation is often preferred because it provides a standardized measure.

### Q11. What is the formula for calculating the sample mean? Provide an example calculation for a dataset.

The formula for the sample mean (\(\bar{x}\)) is:
\[ \bar{x} = \frac{\sum_{i=1}^{n} x_i}{n} \]
where \(x_i\) are the sample values and \(n\) is the number of values in the sample.

Example:
For the dataset \[3, 5, 7, 9\],
\[ \bar{x} = \frac{3 + 5 + 7 + 9}{4} = 6 \]

### Q12. For a normal distribution data what is the relationship between its measure of central tendency?

For a normal distribution, the mean, median, and mode are all equal and lie at the center of the distribution.

### Q13. How is covariance different from correlation?

Covariance indicates the direction of the linear relationship between variables but not the strength. Correlation, on the other hand, standardizes this measure to provide both the direction and the strength of the relationship on a scale from -1 to 1.

### Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

Outliers can significantly affect measures of central tendency and dispersion:
- **Mean**: Outliers can distort the mean, making it higher or lower than the central value of most data.
- **Median**: The median is less affected by outliers because it depends on the middle value(s).
- **Mode**: Outliers do not affect the mode unless they occur frequently.
- **Range**: Outliers increase the range.
- **Variance and Standard Deviation**: Outliers increase both, as they contribute to larger squared deviations from the mean.

Example:
Consider the dataset \[10, 12, 12, 13, 12, 100\]:
- **Mean**: \(\frac{10 + 12 + 12 + 13 + 12 + 100}{6} = 26.5\)
- **Median**: 12
- **Mode**: 12
- **Range**: \(100 - 10 = 90\)
- **Standard Deviation**: Higher due to the outlier (100).
```
