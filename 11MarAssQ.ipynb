```python
### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

**Difference between t-test and z-test**:

- **t-test**:
  - Used when the sample size is small (typically n < 30).
  - Used when the population standard deviation is unknown.
  - Based on the t-distribution, which has heavier tails than the normal distribution.

- **z-test**:
  - Used when the sample size is large (typically n â‰¥ 30).
  - Used when the population standard deviation is known.
  - Based on the normal distribution (z-distribution).

**Example scenarios**:
- **t-test**: Suppose a small startup wants to know if their new employee training program is effective. They have a sample of 15 employees who went through the training, and they want to compare their performance scores to the company's historical average score.
- **z-test**: A researcher wants to determine if a new drug has an effect on blood pressure. They have a large sample of 1000 patients and know the population standard deviation of blood pressure.

### Q2: Differentiate between one-tailed and two-tailed tests.

- **One-tailed test**:
  - Tests for the possibility of an effect in one direction.
  - Hypotheses are directional.
  - Example: \(H_0\): \(\mu \leq 10\), \(H_1\): \(\mu > 10\)

- **Two-tailed test**:
  - Tests for the possibility of an effect in both directions.
  - Hypotheses are non-directional.
  - Example: \(H_0\): \(\mu = 10\), \(H_1\): \(\mu \neq 10\)

### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

- **Type 1 Error** (\(\alpha\)):
  - Incorrectly rejecting the null hypothesis when it is true.
  - Example: A medical test indicates a patient has a disease when they actually do not (false positive).

- **Type 2 Error** (\(\beta\)):
  - Failing to reject the null hypothesis when it is false.
  - Example: A medical test fails to detect a disease in a patient who actually has it (false negative).

### Q4: Explain Bayes's theorem with an example.

**Bayes's Theorem**:
\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]

**Example**:
Suppose a factory produces 1% defective items. If a test correctly identifies defective items 99% of the time (true positive rate) and correctly identifies non-defective items 99% of the time (true negative rate), what is the probability that an item is defective if the test indicates it is defective?

Let:
- \(A\) be the event that an item is defective.
- \(B\) be the event that the test indicates the item is defective.

Given:
- \(P(A) = 0.01\)
- \(P(B|A) = 0.99\)
- \(P(B|\neg A) = 0.01\)
- \(P(\neg A) = 0.99\)

\[
P(B) = P(B|A)P(A) + P(B|\neg A)P(\neg A) = 0.99 \times 0.01 + 0.01 \times 0.99 = 0.0198
\]

\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} = \frac{0.99 \times 0.01}{0.0198} \approx 0.50
\]

### Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

**Confidence Interval**:
A confidence interval (CI) is a range of values used to estimate the true value of a population parameter. It is associated with a confidence level that quantifies the level of confidence that the interval contains the parameter.

**Calculation**:
\[
CI = \bar{x} \pm t \left(\frac{s}{\sqrt{n}}\right)
\]

**Example**:
A sample of 30 students has an average test score of 70 with a standard deviation of 10. Calculate the 95% confidence interval for the true mean.

Given:
- \(\bar{x} = 70\)
- \(s = 10\)
- \(n = 30\)
- \(t_{0.025, 29} \approx 2.045\)

Margin of error:
\[
ME = t \left(\frac{s}{\sqrt{n}}\right) = 2.045 \left(\frac{10}{\sqrt{30}}\right) \approx 3.73
\]

Confidence interval:
\[
70 \pm 3.73 \implies [66.27, 73.73]
\]

### Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.

**Problem**:
A test for a rare disease has a 99% sensitivity (true positive rate) and a 99% specificity (true negative rate). If 0.1% of the population has the disease, what is the probability that a person has the disease given a positive test result?

Let:
- \(A\) be the event that a person has the disease.
- \(B\) be the event that the test is positive.

Given:
- \(P(A) = 0.001\)
- \(P(B|A) = 0.99\)
- \(P(B|\neg A) = 0.01\)
- \(P(\neg A) = 0.999\)

\[
P(B) = P(B|A)P(A) + P(B|\neg A)P(\neg A) = 0.99 \times 0.001 + 0.01 \times 0.999 = 0.01098
\]

\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} = \frac{0.99 \times 0.001}{0.01098} \approx 0.0902
\]

### Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5. Interpret the results.

Given:
- \(\bar{x} = 50\)
- \(s = 5\)
- \(n = 30\)
- \(t_{0.025, 29} \approx 2.045\)

Margin of error:
\[
ME = t \left(\frac{s}{\sqrt{n}}\right) = 2.045 \left(\frac{5}{\sqrt{30}}\right) \approx 1.87
\]

Confidence interval:
\[
50 \pm 1.87 \implies [48.13, 51.87]
\]

**Interpretation**:
We are 95% confident that the true population mean lies between 48.13 and 51.87.

### Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

**Margin of Error (ME)**:
The margin of error is the range added and subtracted from the sample mean to create the confidence interval. It reflects the extent of the expected sampling error.

\[
ME = t \left(\frac{s}{\sqrt{n}}\right)
\]

**Effect of Sample Size**:
As the sample size increases, the margin of error decreases because the term \(\frac{1}{\sqrt{n}}\) gets smaller.

**Example**:
Consider two surveys measuring the same population parameter:
- Survey 1: \(n = 50\), \(s = 5\), \(t_{0.025, 49} \approx 2.0096\)
- Survey 2: \(n = 200\), \(s = 5\), \(t_{0.025, 199} \approx 1.972\)

Margin of error for Survey 1:
\[
ME_1 = 2.0096 \left(\frac{5}{\sqrt{50}}\right) \approx 1.42
\]

Margin of error for Survey 2:
\[
ME_2 = 1.972 \left(\frac{5}{\sqrt{200}}\right) \approx 0.70
\]

### Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

**Z-score Calculation**:
\[
z = \frac{x - \mu}{\sigma} = \frac{75 - 70}{5} = 1
\]

**Interpretation**:
A z-score of 1 indicates that the data point is 1 standard deviation above the mean.

### Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

**

Hypotheses**:
- \(H_0\): \(\mu = 0\) (no weight loss)
- \(H_1\): \(\mu \neq 0\) (weight loss)

Given:
- \(\bar{x} = 6\)
- \(s = 2.5\)
- \(n = 50\)

t-statistic:
\[
t = \frac{\bar{x} - \mu}{s / \sqrt{n}} = \frac{6 - 0}{2.5 / \sqrt{50}} \approx 16.97
\]

Using a t-distribution with 49 degrees of freedom, the critical t-value for a two-tailed test at \(\alpha = 0.05\) is approximately Â±2.0096.

Since 16.97 > 2.0096, we reject \(H_0\). There is significant evidence to support that the drug is effective.

### Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.

Given:
- \(\hat{p} = 0.65\)
- \(n = 500\)
- \(z_{0.025} = 1.96\)

Standard error:
\[
SE = \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}} = \sqrt{\frac{0.65 \times 0.35}{500}} \approx 0.021
\]

Margin of error:
\[
ME = z \times SE = 1.96 \times 0.021 \approx 0.041
\]

Confidence interval:
\[
0.65 \pm 0.041 \implies [0.609, 0.691]
\]

### Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.

**Hypotheses**:
- \(H_0\): \(\mu_1 = \mu_2\)
- \(H_1\): \(\mu_1 \neq \mu_2\)

Given:
- \(\bar{x}_1 = 85, s_1 = 6, n_1 = 30\)
- \(\bar{x}_2 = 82, s_2 = 5, n_2 = 40\)

Pooled standard deviation:
\[
s_p = \sqrt{\frac{(n_1 - 1)s_1^2 + (n_2 - 1)s_2^2}{n_1 + n_2 - 2}} = \sqrt{\frac{(29 \times 36) + (39 \times 25)}{68}} \approx 5.46
\]

t-statistic:
\[
t = \frac{\bar{x}_1 - \bar{x}_2}{s_p \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}} = \frac{85 - 82}{5.46 \sqrt{\frac{1}{30} + \frac{1}{40}}} \approx 2.27
\]

Using a t-distribution with 68 degrees of freedom, the critical t-value for a two-tailed test at \(\alpha = 0.01\) is approximately Â±2.66.

Since 2.27 < 2.66, we fail to reject \(H_0\). There is not enough evidence to support that the teaching methods have a significant difference in student performance.

### Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 90% confidence interval for the true population mean.

Given:
- \(\bar{x} = 65\)
- \(\sigma = 8\)
- \(n = 50\)
- \(z_{0.05} = 1.645\)

Standard error:
\[
SE = \frac{\sigma}{\sqrt{n}} = \frac{8}{\sqrt{50}} \approx 1.13
\]

Margin of error:
\[
ME = z \times SE = 1.645 \times 1.13 \approx 1.86
\]

Confidence interval:
\[
65 \pm 1.86 \implies [63.14, 66.86]
\]

### Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

**Hypotheses**:
- \(H_0\): \(\mu = \text{population reaction time}\)
- \(H_1\): \(\mu \neq \text{population reaction time}\)

Given:
- \(\bar{x} = 0.25\)
- \(s = 0.05\)
- \(n = 30\)
- \(t_{0.05, 29} \approx 1.699\)

t-statistic:
\[
t = \frac{\bar{x} - \mu}{s / \sqrt{n}} = \frac{0.25 - \mu}{0.05 / \sqrt{30}} \approx t
\]

Depending on \(\mu\), compare the t-statistic with the critical value to determine significance. If the calculated t-value is greater than 1.699, reject \(H_0\).
```
